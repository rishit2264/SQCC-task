creating data frame for regression of Last traded price

```python
df1 = df.drop("Company Name",axis = 1)
```

```python
df1 = df1.drop("Symbol",axis = 1)
df1.head()
```

```python
columns_to_del = df[['Percentage Change', '365 Day Percentage Change', '30 Day Percentage Change',"Change"]]
```

```python
df1 = df1.drop(columns_to_del,axis = 1)
df1.head()
```

```python
X = df1.drop("Last Traded Price",axis= 1)
y = df1["Last Traded Price"]
```

scaling the X features.

```python
from sklearn.preprocessing import StandardScaler
se = StandardScaler()
```

```python
X = se.fit_transform(X)
```

using random forest for regression

```python
from sklearn.ensemble import RandomForestRegressor
re = RandomForestRegressor()
```

using k-fold cross val to compare with other models.

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.svm import  SVR
```

```python
re.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
y_pred = re.predict(X_test)
```

calculating MAE(Mean Absolute Error)

```python
mean_absolute_error(y_test,y_pred)
```

calculating mean squared error

```python
mean_squared_error(y_test,y_pred)
```

```python
re.score(X_test,y_test)
```

now we make dataframe for classification of percentage change

```python
df.head()
```

```python
df.columns
```

```python
columns_to_drop = df[["Change","Last Traded Price","Company Name","Symbol","365 Day Percentage Change","30 Day Percentage Change"]]
```

```python
df2 = df.drop(columns_to_drop,axis = 1)
df2.head()
```

here we use pandas to numeric as when i was using directly lambda it was showing percentage change has some string values.

```python
df2['Percentage Change'] = pd.to_numeric(df2['Percentage Change'], errors='coerce')
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
df2.head()
```

Converting Percentage Change into a binary target (positive or negative).

```python
df2["Percentage Change"] = df2["Percentage Change"].apply(lambda x: 1 if x>0 else 0)
```

```python
X1 = df2.drop("Percentage Change",axis = 1)
y1 = df2["Percentage Change"]
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size =0.2)
```

```python
rfc.fit(X1_train,y1_train)
```

```python
rfc.score(X1_test,y1_test)
```
